# Restful API & Flask

1. What is RESTFUl API ?
   - A RESTful API for data science is simply a REST API used to serve, access, or interact with data science workflows and models over the web.


2. What is Flask, and why is it popular for building APIs?
   - Flask is a lightweight web framework in Python used to build web applications and APIs quickly and with minimal setup. It’s often called a “micro-framework” because it provides just the essentials—routing, request handling, and response handling—without forcing a complex structure.

     It is popular for building APIs beacuse of the following reasons:
     - Simple and easy to learn
     - Lightweight and flexible
     - Perfect for data science
     - Fast to build and deploy
     - Huge ecosystem



3. What are HTTP methods used in RESTful APIs?
   - HTTP methods (also called verbs) define the type of action a client wants to perform on a resource in a RESTful API.
      - | Method | Purpose          | Example Endpoint |
| ------ | ---------------- | ---------------- |
| GET    | Read data        | `/users`         |
| POST   | Create new data  | `/users`         |
| PUT    | Update (full)    | `/users/1`       |
| PATCH  | Update (partial) | `/users/1`       |
| DELETE | Remove data      | `/users/1`       |




4.  What is the purpose of the @app.route() decorator in Flask?
    - The @app.route() decorator in Flask is used to connect a URL (endpoint) to a Python function. In simple terms, it tells Flask:“When someone visits this URL, run this function.”


5. What is the role of Flask-SQLAlchemy?
   - Flask-SQLAlchemy is an extension for Flask that makes it easier to work with databases using Python code instead of raw SQL.It is built on top of SQLAlchemy, a powerful ORM (Object Relational Mapper).
